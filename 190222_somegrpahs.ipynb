# Seaborn, Matplot visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rc('ytick',labelsize=15)
# To use 한글 font in graphs
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus']=False
# 그래프에서 음수값이 나올 때, 깨지는 현상 방지
plt.rc('axes', unicode_minus = False)
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'http://bitly.kr/yKYiM'
data1 = pd.read_csv(url)
```

```python
data1.describe(exclude='object')
```

# Countplot

```python
sns.countplot(data1['MDC'])
```

```python
sns.set_style('darkgrid') 
sns.countplot(data1['MDC'])
```

* 'whitegrid', 'darkgrid', 'dark', 'white', 'ticks'

```python
plt.figure(figsize=(14,7))
sns.set_style('darkgrid',) 
sns.countplot(data = data1, x='race', hue='problem')
```

```python
sns.palplot(sns.color_palette('deep',10))
sns.palplot(sns.color_palette('colorblind',10))
sns.palplot(sns.color_palette('gray',10))
```

```python
plt.figure(figsize=(14,7))
sns.set_style('darkgrid',) 
sns.countplot(data = data1, x='race', hue='problem', color='g',
              palette=sns.color_palette('deep'))
```

```python
# 글씨가 그래프 사이에서 겹칠 때
plt.tight_layout()
```

# distplot

```python
sns.distplot(data1['lat'], hist=False)
```

```python
sns.distplot(data1['lat'], kde=False)
```

```python
sns.distplot(data1['lat'], kde=False, vertical=True)
```

```python
one = data1[data1['gender'] == 'Male']
two = data1[data1['gender']== 'Female']
sns.distplot(one['lat'], label='남성', hist=False)
sns.distplot(two['lat'], label='여성', hist=False)
plt.tight_layout()
plt.legend()
```

# Bar plot

```python
plt.figure(figsize=(12,6))
plt.ylim([44.88, 45])
sns.barplot(data=data1, x='race', y='lat',hue='problem', ci ='sd')
plt.tight_layout()
```

```python
plt.figure(figsize=(12,6))
# plt.ylim([44.88, 45])
sns.barplot(data=data1, x='race', y='lat', estimator=sum)
plt.tight_layout()
```

# Line plot / Point plot

```python
sns.pointplot(data=data1, x='policePrecinct', y='lat', hue='problem')
plt.xticks(size=20)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.describe(exclude='object')
```

```python
sns.lineplot(data=fmri, x='timepoint', y= 'signal', hue='event')
```

# Box plot

```python
sns.boxplot(data=data1, x='lat')
```

```python
sns.boxplot(data=data1, x='race', y='lat')
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(data=data1, x='race', y='lat', hue='gender')
plt.tight_layout()
```

# Violin plot

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=data1, x='race', y='lat')
```
